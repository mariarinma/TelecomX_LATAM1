```python
def churn_por_categoria(columna):
    tabla = df.groupby(columna)["Churn"].mean().sort_values(ascending=False)

    print(f"\nTasa de churn por {columna}")
    display((tabla * 100).round(2))

    plt.figure()
    plt.bar(tabla.index.astype(str), tabla.values)
    plt.title(f"Tasa de Churn por {columna}")
    plt.ylabel("Tasa de Churn")
    plt.xticks(rotation=45, ha="right")
    plt.show()
```

# Telecom X – Análisis de Evasión de Clientes (Churn)

## Proyecto de Análisis Exploratorio de Datos (EDA)

Este proyecto tiene como objetivo analizar los factores asociados a la evasión de clientes (Churn) en Telecom X, identificando patrones y variables que influyen en la cancelación del servicio.

## 1. Introducción

La evasión de clientes (Churn) representa uno de los principales desafíos para empresas de telecomunicaciones, ya que implica pérdida de ingresos, aumento en costos de adquisición de nuevos clientes y menor estabilidad financiera.

El objetivo de este análisis es:

- Comprender la magnitud del churn.
- Identificar variables asociadas con la cancelación.
- Detectar patrones en perfiles de clientes.
- Generar insights estratégicos que permitan diseñar acciones de retención.

Se trabajó con un dataset de 7,043 clientes activos y cancelados, incluyendo información demográfica, contractual y financiera.

## 2. Limpieza y Tratamiento de Datos

Durante el proceso de preparación se realizaron las siguientes acciones:

- Eliminación de 224 registros con valores vacíos en la variable objetivo `Churn`.
- Conversión de `account.Charges.Total` a formato numérico.
- Reemplazo de valores faltantes en `account.Charges.Total` por 0 (clientes con antigüedad 0).
- Conversión de la variable `Churn` a formato binario (0 = No canceló, 1 = Canceló).
- Transformación de variables binarias (Yes/No) a valores numéricos.
- Corrección de valores como "No internet service" antes de su transformación.
- Creación de la variable derivada `Cuentas_Diarias`, calculada como el cargo mensual dividido entre 30 días.

Al finalizar, el dataset quedó compuesto por 7,043 registros completos y consistentes.

## 3. Análisis Exploratorio de Datos

### - Distribución del Churn

Se identificó que el 26.54% de los clientes cancelaron el servicio, mientras que el 73.46% permanecen activos.

Esto indica que aproximadamente 1 de cada 4 clientes abandona la compañía, lo que representa un nivel de evasión significativo.

---

### - Churn por Tipo de Contrato

- Month-to-month: 42.71%
- One year: 11.27%
- Two year: 2.83%

Los contratos mensuales presentan una tasa de evasión considerablemente mayor. La duración contractual es uno de los factores más determinantes en la retención.

---

### - Churn por Método de Pago

- Electronic check: 45.29%
- Mailed check: 19.11%
- Bank transfer (automatic): 16.71%
- Credit card (automatic): 15.24%

Los clientes con métodos de pago automáticos presentan menor probabilidad de cancelación.

---

### - Variables Numéricas y Churn

Antigüedad promedio:
- No Churn: 37.57 meses
- Churn: 17.98 meses

Los clientes que cancelan tienen menor permanencia en la empresa.

Cargo mensual promedio:
- No Churn: 61.27
- Churn: 74.44

Los clientes con mayor cargo mensual presentan mayor probabilidad de evasión.

En conjunto, la antigüedad y el nivel de cargo mensual son variables clave asociadas al churn.

## 4. Conclusiones e Insights

Los principales factores asociados a la evasión son:

1. Tipo de contrato (principal factor determinante).
2. Antigüedad del cliente.
3. Método de pago.
4. Nivel de cargo mensual y diario.

Se observa que:
- Los clientes con contratos mensuales son altamente propensos a cancelar.
- La evasión ocurre principalmente en los primeros meses.
- Los clientes con planes más costosos presentan mayor riesgo.
- Los métodos de pago automáticos están asociados con mayor estabilidad.

La combinación de estos factores permite identificar segmentos de alto riesgo.

## 5. Recomendaciones Estratégicas

Con base en los hallazgos, se proponen las siguientes acciones:

- Incentivar contratos a largo plazo  
Ofrecer descuentos o beneficios para migrar clientes de contratos mensuales a contratos anuales o bianuales.

- Programas de retención temprana  
Implementar acciones de seguimiento durante los primeros 12 meses, etapa donde se concentra mayor evasión.

- Promover pagos automáticos  
Ofrecer incentivos para clientes que adopten métodos de pago automáticos.

- Revisar estructura de precios  
Analizar la percepción de valor en planes de mayor costo, ya que presentan mayor tasa de cancelación.

- Segmentación predictiva  
Utilizar estas variables como base para modelos de predicción de churn que permitan intervenir antes de la cancelación.

La implementación de estas estrategias podría reducir significativamente la tasa de evasión y mejorar la estabilidad financiera de la empresa.
